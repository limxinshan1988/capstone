# Coordinates of Singapore areas

```python
!pip install geocoder
```

```python
import geocoder
```

```python
def get_latlng(neighborhood):
    lat_lng_coords = None
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, Singapore'.format(neighborhood))
        lat_lng_coords = g.latlng
    return lat_lng_coords
```

```python
coords = [ get_latlng(neighborhood) for neighborhood in SG_df["SG_Area"].tolist() ]
```

```python
radius = 5000
LIMIT = 100

venues = []

for lat, long, neighborhood in zip(SG_df6['Latitude'], SG_df6['Longitude'], SG_df6['SG_Area']):
    
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        Client_ID,
        Client_secret,
        Version,
        lat,
        long,
        radius, 
        LIMIT)
    
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    for venue in results:
        venues.append((
            neighborhood,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

# K means clustering

```python
kclusters = 3

SG_cluster = SG_Gym.drop(["SG_Area"], 1)

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(SG_cluster)

kmeans.labels_[0:10] 
```

```python
SG_merged = SG_Gym.copy()

SG_merged["Cluster Labels"] = kmeans.labels_
```

```python
SG_merged.rename(columns={"SG_Area": "SG_Area"}, inplace=True)
SG_merged.head()
```

# Cluster visualization

```python
map_SG = folium.Map(location=[latitude, longitude], zoom_start=12)

x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, poi, cluster in zip(SG_merged['Latitude'], SG_merged['Longitude'], SG_merged['SG_Area'], SG_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=8,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_SG)
       
map_SG
```
